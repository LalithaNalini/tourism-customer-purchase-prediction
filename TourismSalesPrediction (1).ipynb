```python
from datasets import load_dataset

# Load the train and test datasets from the Hugging Face Hub
# Replace "LalithaShiva/tourism_prediction_dataset" with your actual dataset repository ID
try:
    dataset = load_dataset("LalithaShiva/tourism_prediction_dataset", split=['train', 'test'])
    train_dataset = dataset[0]
    test_dataset = dataset[1]
    print("Train and test datasets loaded successfully!")
    print("Train dataset:", train_dataset)
    print("Test dataset:", test_dataset)
except Exception as e:
    print(f"Error loading datasets: {e}")
    train_dataset = None
    test_dataset = None
```

```python
from sklearn.ensemble import RandomForestClassifier

# Define the RandomForestClassifier model
model = RandomForestClassifier(random_state=42)

# Define parameters for tuning
parameters = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30]
}

print("RandomForestClassifier model and parameters defined.")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Assuming your features are in 'train_dataset' and the target is 'ProdTaken'
# You might need to adjust the column names based on your dataset

# Convert datasets to pandas DataFrames for easier handling with scikit-learn
train_df = train_dataset.to_pandas()
test_df = test_dataset.to_pandas()

# Separate features (X) and target (y)
X_train = train_df.drop("ProdTaken", axis=1)
y_train = train_df["ProdTaken"]
X_test = test_df.drop("ProdTaken", axis=1)
y_test = test_df["ProdTaken"]

# Handle categorical features and potential missing values
# This is a basic example, you might need more sophisticated preprocessing
for col in X_train.columns:
    if X_train[col].dtype == 'object':
        # Simple imputation for demonstration
        X_train[col].fillna(X_train[col].mode()[0], inplace=True)
        X_test[col].fillna(X_test[col].mode()[0], inplace=True)

        # Encode categorical features
        le = LabelEncoder()
        X_train[col] = le.fit_transform(X_train[col])
        X_test[col] = le.transform(X_test[col])
    else:
        # Simple imputation for numerical features
        X_train[col].fillna(X_train[col].median(), inplace=True)
        X_test[col].fillna(X_test[col].median(), inplace=True)


# Initialize GridSearchCV
# 'model' and 'parameters' are assumed to be defined in the previous cell
grid_search = GridSearchCV(estimator=model, param_grid=parameters, cv=3, n_jobs=-1, verbose=2)

# Perform the grid search on the training data
print("Starting grid search...")
grid_search.fit(X_train, y_train)

# Get the best parameters and best score
best_parameters = grid_search.best_params_
best_score = grid_search.best_score_

print("\nGrid Search completed.")
print("Best parameters found:", best_parameters)
print("Best cross-validation score:", best_score)

# The best model is available in grid_search.best_estimator_
best_model = grid_search.best_estimator_
```

```python
# Log the tuned parameters
print("\nTuned Parameters:")
for param, value in best_parameters.items():
    print(f"- {param}: {value}")

# You can also log these parameters to a file or an experiment tracking tool if you are using one.
# Example using a simple file write:
with open("tuned_parameters.txt", "w") as f:
    f.write("Tuned Parameters:\n")
    for param, value in best_parameters.items():
        f.write(f"- {param}: {value}\n")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Evaluate the best model on the test data
# 'best_model' is assumed to be available from the previous grid search step

# Make predictions on the test set
y_pred = best_model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("\nModel Evaluation on Test Set:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# You can add other relevant metrics or visualizations here.
```
